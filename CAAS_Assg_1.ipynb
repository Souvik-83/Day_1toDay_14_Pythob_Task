1. What is difference between shallow copy and deep copy? 

The difference between shallow copy and deep copy lies in how they create copies of objects containing other objects (such as lists or dictionaries) in Python.

Shallow copy:

A shallow copy creates a new object but references the same nested objects as the original. In other words, it creates a new container object but does not duplicate the elements inside it.
Modifying the nested objects in the copy will affect the original object, as they both refer to the same underlying objects.
Shallow copy is performed using the copy() method or the built-in copy module's copy() function.
Deep copy:

A deep copy creates a new object and recursively duplicates all the nested objects within it. It creates entirely independent copies of the original object and its nested objects.
Modifying the nested objects in the deep copy will not affect the original object or its nested objects, as they are completely separate.
Deep copy is performed using the deepcopy() method from the copy module.

3. What is duck typing in python, what are Dunder or special methods?  

Duck typing is a concept in Python that focuses on the behavior of objects rather than their specific types. It follows the principle: "If it looks like a duck, swims like a duck, and quacks like a duck, then it probably is a duck." In other words, Python determines the suitability of an object based on its compatibility with the required methods and attributes, rather than its actual type or class inheritance.
Dunder (double underscore) methods, also known as special methods or magic methods, are a set of predefined methods in Python that enable customized behavior for classes. These methods are surrounded by double underscores (e.g., __init__, __str__, __add__).
